```python
df_re_mg_sku_order=df_mg_sku_order.rename(columns={"Order Qty":"Order_Qty","Weight (g)":"Weight_g"})
```

```python
convert_g_to_kg="""Total_Weight_Per_Shipment=(Order_Qty*Weight_g)/@kg"""
df_re_mg_sku_order.eval(convert_g_to_kg,inplace=True,engine='python')
```

```python
re_df_mg_sku_order=df_re_mg_sku_order.rename(columns={"Order_Qty":"Order Qty","Total_Weight_Per_Shipment":"Total Weight As Per X (kg)","Weight_g":"Weight (g)"})
re_df_mg_sku_order.sample()
```

```python
def weight_slab(weight):
    i=round(weight%1,1)
    if i==0.0:
        return weight
    elif i>0.5:
        return int(weight)+1.0
    else:
        return int(weight)+0.5
```

```python
re_df_mg_sku_order['Weight Slab As Per X (kg)']=re_df_mg_sku_order['Total Weight As Per X (kg)'].apply(weight_slab)
df_courier_invoice['Weight Slab Charged By Courier Company (kg)']=(df_courier_invoice['Total Weight As Per Courier Company(kg)']).apply(weight_slab)
```

```python
re_df_mg_sku_order
```

```python
total_Expected_charge=[]
for i in range(len(re_df_mg_sku_order)):
    fwd_category='fwd_'+re_df_mg_sku_order['Delivery Zone As Per X'][i]
    fwd_fixed=df_courier_company_rates[fwd_category+'_fixed'][0]
    fwd_additional=df_courier_company_rates[fwd_category+'_additional'][0]
    rto_category='rto_'+re_df_mg_sku_order['Delivery Zone As Per X'][i]
    rto_fixed=df_courier_company_rates[rto_category+'_fixed'][0]
    rto_additional=df_courier_company_rates[rto_category+'_additional'][0]
    if re_df_mg_sku_order['Type of Shipment'][i]=='Forward charges':
        if re_df_mg_sku_order['Weight Slab As Per X (kg)'][i]<=0.5:
            total_Expected_charge.append(fwd_fixed)
        else:
            additional_weight=(re_df_mg_sku_order['Weight Slab As Per X (kg)'][i]-0.5)/0.5
            total_Expected_charge.append(fwd_fixed+additional_weight*fwd_additional)
    if re_df_mg_sku_order['Type of Shipment'][i]=='Forward and RTO charges':
        if re_df_mg_sku_order['Weight Slab As Per X (kg)'][i]<=0.5:
            total_Expected_charge.append(fwd_fixed+rto_fixed)
        else:
            additional_weight=(re_df_mg_sku_order['Weight Slab As Per X (kg)'][i]-0.5)/0.5
            total_Expected_charge.append(fwd_fixed+additional_weight*(fwd_additional+rto_additional))
            
re_df_mg_sku_order['Expected Charge As Per X (Rs.)']=total_Expected_charge
```

```python
Output2=re_df_mg_sku_order[['Order ID',
       'Type of Shipment','Delivery Zone As Per X',
       'Total Weight As Per X (kg)', 'Weight Slab As Per X (kg)',
       'Expected Charge As Per X (Rs.)']]
```

```python
Output1=df_courier_invoice[['AWB Code', 'Order ID', 'Total Weight As Per Courier Company(kg)',
       'Delivery Zone Charged By Courier Company','Billing Amount (Rs.)', 'Weight Slab Charged By Courier Company (kg)']]
```

```python
df_merged_output=Output2.merge(Output1,on='Order ID',how='left')
df_merged_output.head(5)
```

```python
resultant_output1=df_merged_output.copy()
```

```python
resultant_output1=resultant_output1.rename(columns={'AWB Code':'AWB Number','Billing Amount (Rs.)':'Charges_Billed_By_Courier_Company_Rs','Expected Charge As Per X (Rs.)':'Expected_Charge_As_Per_X_Rs'})
resultant_output1.head()
```

```python
diff_bw_ec_bc="""
Difference_Between_Expected_Charges_and_Billed_Charges_Rs =(Expected_Charge_As_Per_X_Rs - Charges_Billed_By_Courier_Company_Rs)
"""
resultant_output1.eval(diff_bw_ec_bc,inplace=True,engine='python')
```

```python
resultant_output1=resultant_output1.rename(columns={'Charges_Billed_By_Courier_Company_Rs':'Charges Billed By Courier Company (Rs.)','Expected_Charge_As_Per_X_Rs':'Expected Charge As Per X (Rs.)','Difference_Between_Expected_Charges_and_Billed_Charges_Rs':'Difference Between Expected Charges And Billed Charges (Rs.)'})
```

```python
writer = pd.ExcelWriter('Desktop/CoinTab/Expected_Result.xlsx',if_sheet_exists="replace",engine='openpyxl',mode="a")
```

```python
resultant_output1.to_excel(writer,sheet_name='Calculations')
calculations=writer.save()
```

```python
correctly_charged=resultant_output1[resultant_output1['Difference Between Expected Charges And Billed Charges (Rs.)']==0.0]
correctly=["Total Orders where X has been correctly charged",len(correctly_charged),sum(correctly_charged['Charges Billed By Courier Company (Rs.)'])]

over_charged=resultant_output1[resultant_output1['Difference Between Expected Charges And Billed Charges (Rs.)']<0.0]
over=["Total Orders where X has been over charged",len(over_charged),np.absolute(sum(over_charged['Difference Between Expected Charges And Billed Charges (Rs.)']))]

under_charged=resultant_output1[resultant_output1['Difference Between Expected Charges And Billed Charges (Rs.)']>0.0]
under=["Total Orders where X has been under charged",len(under_charged),np.absolute(sum(under_charged['Difference Between Expected Charges And Billed Charges (Rs.)']))]
```

```python
resultant_output2=pd.DataFrame([correctly,over,under],columns=['Description','Count','Amount (Rs.)'])
```
